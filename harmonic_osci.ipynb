# Demo for Harmonic Oscillator potential
### The 1D Hamiltonian is solved numerically for a quadratic potential. Results are compared with the exact solutions.

```python
from _ham import Hamiltonian
```

```python
def harmonic_potential(x):
    return 0.5 * x**2
```

```python
ham_hamosc = Hamiltonian(20,201)
```

```python

ham_hamosc.build_hamiltonian(harmonic_potential)
```

```python
ham_hamosc.diagonalize()
```

## Results


### Eigenvalues: Exact (blue) vs. numeric (red)

```python
import matplotlib.pyplot as plt
plt.style.use('scientific.mplstyle')
plt.figure(figsize=(10,8))
plt.plot(ham_hamosc.x_values, harmonic_potential(ham_hamosc.x_values), color='black', zorder=3)
for i in range(70):
    plt.axhline(i + .5)
    plt.axhline(ham_hamosc.eigvals[i], color='red', linestyle='dashed')
plt.xlim(-8,8)
```
