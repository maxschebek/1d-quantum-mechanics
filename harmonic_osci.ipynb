# Demo for Harmonic Oscillator potential
### The 1D Hamiltonian is solved numerically for a quadratic potential. Results are compared with the exact solutions.

```python
from _ham import Hamiltonian
```

```python
def harmonic_potential(x):
    return 0.5 * x**2
```

```python
# Inititalize Hamiltonian
length = 20
n_steps = 201
ham_harmonic = Hamiltonian(length,201)
```

```python

ham_harmonic.build_hamiltonian(harmonic_potential)
```

```python
ham_harmonic.diagonalize()
```

## Results


### Eigenvalues: Exact (blue) vs. numeric (red)

```python
import matplotlib.pyplot as plt
plt.style.use('scientific.mplstyle')
plt.figure(figsize=(8,6))
plt.plot(ham_harmonic.x_values, harmonic_potential(ham_harmonic.x_values), color='black', zorder=3)
for i in range(70):
    plt.axhline(i + .5)
    plt.axhline(ham_harmonic.eigvals[i], color='red', linestyle='dashed')
plt.xlim(-8,8)
```

###  Square modulus of the 10 lowest-lying states (shifted by correspnding energy value and scaled by factor 10)

```python
import matplotlib.pyplot as plt
plt.style.use('scientific.mplstyle')
plt.figure(figsize=(8,6))
for i in range(10):
    plt.plot( ham_harmonic.x_values,ham_harmonic.eigvals[i]+10*ham_harmonic.eigvecs[:,i]**2)
# plt.plot(ham_harmonic.x_values,ham_harmonic.eigvecs[:,0]**2)
# plt.plot(ham_harmonic.x_values,ham_theta.eigvals[2] + ham_theta.eigvecs[:,2]**2)
plt.plot(ham_harmonic.x_values, harmonic_potential(ham_harmonic.x_values), color='black')

plt.xlabel('$x$'), plt.ylabel('$|\Psi(x)|^2$')
plt.ylim(0,10)
plt.xlim(-8,8)

plt.show()
```
